## Model for important features
Trước hết ta sẽ đưa tất cả các cột dữ liệu vào trong các mô hình phân lớp, cái ta quan tâm ở đây đó chính là mức độ tương quan các cột đối với target colunm ở đây là smoking colunm, từ đó đánh giá được mức độ ảnh hưởng khác nhau và lựa chọn cột có ảnh hướng lớn tới dự đoán.\
Giải thích phương pháp thực hiện ở đây đó chính là:\
Chúng tôi đặt câu hỏi ở đây đó chính là nếu như thay vì ta xáo(giống kiểu xào bài) một cột bất kì trong tập dữ liệu và giữ nguyên đối với tập target(ở đây là cột smoking) và cũng như các cột khác, điều đó liệu sẽ ảnh hưởng như thế nào với dự đoán ? Điều này sẽ giúp dễ dàng đánh giá tầm quan trọng của mô hình.\

Các bước thực hiện:\
1. Thực hiện điều này trong tập train
2. Trộn từng cột trong tập dữ liệu trừ tập target(ở đây là tập smoking), sau đó thực hiện dự đoán. Lưu ý độ quan trọng ở đây sẽ sử dụng là Accuracy
3. Điểm important được tính là:\
$$i = origin\;score - avg(shuffer\;score)$$

note: shuffer score là tính toán cho tất cả số lần tráo khác nhau trong tập dữ liệu

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
from eli5.sklearn import PermutationImportance

"""
need plot for visual features important
3 model for predict get their own important data
avg that data for evaluate
"""
def imp_features(model, X_test, y_test):
    perm = PermutationImportance(model, random_state=42).fit(X_test, y_test)
    importances = perm.feature_importances_
    feature_names = X.columns

    feature_importances_df = pd.DataFrame({
        'Feature': feature_names,
        'Importance': importances
    })

    feature_importances_df = feature_importances_df.sort_values(by='Importance', ascending=False)

    plt.figure(figsize=(10, 8))
    sns.barplot(x='Importance', y='Feature', data=feature_importances_df)
    plt.title('Feature Importances for Smoking Prediction')
    plt.show()

def evaluate_model(model, X_test, y_test):
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print("Accuracy:", accuracy)
    print("Report:")
    print(classification_report(y_test, y_pred))

models = {
    'XGBoost': XGBClassifier(random_state=42),
    'Random Forest': RandomForestClassifier(random_state=42),
    'Logistic Regression': LogisticRegression(random_state=42),
}

feature_importance_summary = pd.DataFrame(columns=X.columns)

for name, model in models.items():
    model.fit(X_train, y_train)
    print(f"Evaluating {name}...")
    imp_features(model, X_test, y_test)
    evaluate_model(model, X_test, y_test)
    perm = PermutationImportance(model, random_state=42).fit(X_test, y_test)
    feature_importance_summary.loc[name] = perm.feature_importances_

average_feature_importance = feature_importance_summary.mean()

ranked_features = average_feature_importance.sort_values(ascending=False)

plt.figure(figsize=(10, 8))
sns.barplot(x=ranked_features.values, y=ranked_features.index)
plt.title('Average Feature Importances across Models')
plt.show()
```

# Tính toán BMI ở người
Ở trên có đề cập đến 2 chỉ số heigth và weight để tiện thế nên ta có gộp chúng lại là một chỉ số duy nhất đó chúng là bmi. Bmi là là một cách nhận định cơ thể của một người là gầy hay béo bằng một chỉ số.


```python
def BMI(height, weight):
  return weight / ((height/100) * (height/100))
```

```python
bmiNums, bmis, hemo, tri, gtp = [[] for _ in range(5)]

df = smoke.copy()

def categorize_bmi(rbmi):
  if rbmi < 18.5:
    return 'underweight'
  elif rbmi < 25:
    return 'healthy'
  elif rbmi < 30:
    return 'overweight'
  else:
    return 'obese'


def categorize_hemo(hemoglobin):
  if hemoglobin < 13.2:
    return 'low'
  elif hemoglobin > 16.5:
    return 'high'
  else:
    return 'normal'

def cateforize_tri(triglyceride):
  return 'high' if triglyceride >=150 else 'normal'


# def categorize_fbs(fbs):
#   if fbs <=99:
#     return 'normal'
#   elif fbs <=123:
#     return 'prediabetic'
#   else:
#     return 'diabetic'

def categorize_gtp(GTP):
  return 'normal' if GTP <= 50 else 'high'


for _, record in df.iterrows():
  rbmi = BMI(record['height(cm)'],record['weight(kg)'])
  bmiNums.append(rbmi)
  bmis.append(categorize_bmi(rbmi))
  hemo.append(categorize_hemo(record['hemoglobin']))
  tri.append(cateforize_tri(record['triglyceride']))
  gtp.append(categorize_gtp(record['Gtp']))
  # fbs.append(categorize_fbs(record['fasting blood sugar']))


```

Thêm các trường trên vào dữ liệu

```python
df['BMI'] = bmiNums
df['BMI_Assignment'] = bmis
df['Hemoglobin_Assignment'] = hemo
df['Triglyceride_Assignment'] = tri
df['GTP_Assignment'] = gtp
```

```python
import matplotlib.pyplot as plt
import numpy as np

from matplotlib.patches import Circle, RegularPolygon
from matplotlib.path import Path
from matplotlib.projections import register_projection
from matplotlib.projections.polar import PolarAxes
from matplotlib.spines import Spine
from matplotlib.transforms import Affine2D


def radar_factory(num_vars, frame='circle'):
    """
    Create a radar chart with `num_vars` Axes.

    This function creates a RadarAxes projection and registers it.

    Parameters
    ----------
    num_vars : int
        Number of variables for radar chart.
    frame : {'circle', 'polygon'}
        Shape of frame surrounding Axes.

    """
    # calculate evenly-spaced axis angles
    theta = np.linspace(0, 2*np.pi, num_vars, endpoint=False)

    class RadarTransform(PolarAxes.PolarTransform):

        def transform_path_non_affine(self, path):
            # Paths with non-unit interpolation steps correspond to gridlines,
            # in which case we force interpolation (to defeat PolarTransform's
            # autoconversion to circular arcs).
            if path._interpolation_steps > 1:
                path = path.interpolated(num_vars)
            return Path(self.transform(path.vertices), path.codes)

    class RadarAxes(PolarAxes):

        name = 'radar'
        PolarTransform = RadarTransform

        def __init__(self, *args, **kwargs):
            super().__init__(*args, **kwargs)
            # rotate plot such that the first axis is at the top
            self.set_theta_zero_location('N')

        def fill(self, *args, closed=True, **kwargs):
            """Override fill so that line is closed by default"""
            return super().fill(closed=closed, *args, **kwargs)

        def plot(self, *args, **kwargs):
            """Override plot so that line is closed by default"""
            lines = super().plot(*args, **kwargs)
            for line in lines:
                self._close_line(line)

        def _close_line(self, line):
            x, y = line.get_data()
            # FIXME: markers at x[0], y[0] get doubled-up
            if x[0] != x[-1]:
                x = np.append(x, x[0])
                y = np.append(y, y[0])
                line.set_data(x, y)

        def set_varlabels(self, labels):
            self.set_thetagrids(np.degrees(theta), labels)

        def _gen_axes_patch(self):
            # The Axes patch must be centered at (0.5, 0.5) and of radius 0.5
            # in axes coordinates.
            if frame == 'circle':
                return Circle((0.5, 0.5), 0.5)
            elif frame == 'polygon':
                return RegularPolygon((0.5, 0.5), num_vars,
                                      radius=.5, edgecolor="k")
            else:
                raise ValueError("Unknown value for 'frame': %s" % frame)

        def _gen_axes_spines(self):
            if frame == 'circle':
                return super()._gen_axes_spines()
            elif frame == 'polygon':
                # spine_type must be 'left'/'right'/'top'/'bottom'/'circle'.
                spine = Spine(axes=self,
                              spine_type='circle',
                              path=Path.unit_regular_polygon(num_vars))
                # unit_regular_polygon gives a polygon of radius 1 centered at
                # (0, 0) but we want a polygon of radius 0.5 centered at (0.5,
                # 0.5) in axes coordinates.
                spine.set_transform(Affine2D().scale(.5).translate(.5, .5)
                                    + self.transAxes)
                return {'polar': spine}
            else:
                raise ValueError("Unknown value for 'frame': %s" % frame)

    register_projection(RadarAxes)
    return theta
```
